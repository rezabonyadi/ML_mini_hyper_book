```python
from scipy.optimize import linprog
class lp_sparse_svm():
    def __init__(self):
        pass
    
    def get_mats(self, x, y, p_w, lamb):
        m, n = x.shape 
        lamb = lamb/m
        
        # w, t, epsilons, b
        f = np.squeeze(np.concatenate([np.zeros((n,1)), np.ones((n,1)), lamb*np.ones((m,1)), [[0]]]))
        main_constraints = np.concatenate([np.diag(-y).dot(x), np.zeros((m,n)), -np.eye(m), -y.reshape((-1,1))], axis=1)
        w_less_t = np.concatenate([np.eye(n),-np.eye(n), np.zeros((n,m)), np.zeros((n,1))], axis=1)
        minus_w_less_t = np.concatenate([-np.eye(n),-np.eye(n), np.zeros((n,m)), np.zeros((n,1))], axis=1)

        A = np.concatenate([main_constraints, w_less_t, minus_w_less_t])
        b = np.squeeze(np.concatenate([-np.ones((m,1)), np.zeros((2*n,1))]))

        n_prev_dim = p_w.shape[0]
        
        if n_prev_dim!=0:
            A_eq = np.concatenate([p_w, np.zeros((n_prev_dim, n)), np.zeros((n_prev_dim, m)), np.zeros((n_prev_dim,1))], axis=1)
            b_eq = np.zeros((n_prev_dim))
        else:
            A_eq = []
            b_eq = []
            
        bounds = np.zeros((2, 2*n+1+m))
        bounds[0,:] = np.concatenate([n*[-np.inf], n*[-np.inf], m*[0], [-np.inf]])
        bounds[1,:] = np.concatenate([n*[np.inf], n*[np.inf], m*[np.inf], [np.inf]])    
        bounds = tuple(map(tuple, bounds.T))
        
        return f, A, b, bounds, A_eq, b_eq
    
    def fit(self, x, y, lamb=0.5, p_w=np.empty(0), gamma=.1):
        c, A, b, bounds, A_eq, b_eq = self.get_mats(x, y, p_w, lamb)
        if p_w.shape[0]==0:
            res = linprog(c, A_ub=A, b_ub=b, bounds=bounds, options={'tol':1e-8, 'maxiter':1000, "disp": True})
        else:
            res = linprog(c, A_ub=A, b_ub=b, A_eq=A_eq, b_eq=b_eq, 
                          bounds=bounds, options={'tol': 1e-8, 'maxiter':1000, "disp": True})
            print(A_eq.dot(res.x))
#         print(res)
        m, n = x.shape
        self.w = res.x[0:n]
        self.b = res.x[-1]
        self.A = A
        self.c = c
        self.A_eq = A_eq
        self.b_eq = b_eq
        self.res = res
    
    def predict(self, x):
        return 2*(x.dot(self.w)+self.b>0)-1
```

```python
class lp_svm_dr:
    def __init__(self):
        pass
    
    def transform(self, x, nc=1):
        r = np.zeros((x.shape[0], nc))
        for i in range(nc):  
            x=x-self.means[i,:]
            r[:, i] = x.dot(self.reduction_vector[:,i])+self.intercepts[i]
            
        return r
        
    def fit(self, x, y, nc=1, C=0.5):
        self.reduction_vector = np.zeros((x.shape[1], nc))
        self.intercepts = np.zeros(nc)
        self.means = np.zeros((nc, x.shape[1]))
        
        x1 = x    
        
        p_w = np.zeros((nc-1, x.shape[1]))
        all_weights = []
        self.models = []
        for i in range(nc):  
            print(i)
            l = lp_sparse_svm()
            if i==0:
                l.fit(x, y)
            else:
                l.fit(x, y, p_w=np.asarray(all_weights))
#                 print(l.A_eq.dot(np.asarray(all_weights).T))
                
            self.reduction_vector[:,i] = l.w
            all_weights.append(l.w)
            self.intercepts[i] = l.b     
            self.models.append(l)
```

```python
# import PuLP
from pulp import *

# Create the 'prob' variable to contain the problem data
prob = LpProblem("The Miracle Worker", LpMaximize)

# Create problem variables
x=LpVariable("Medicine_1_units",0,None,LpInteger)
y=LpVariable("Medicine_2_units",0, None, LpInteger)

# The objective function is added to 'prob' first
prob += 25*x + 20*y, "Health restored; to be maximized"
# The two constraints are entered
prob += 3*x + 4*y <= 25, "Herb A constraint"
prob += 2*x + y <= 10, "Herb B constraint"

# The problem data is written to an .lp file
prob.writeLP("MiracleWorker.lp")

# The problem is solved using PuLP's choice of Solver
prob.solve()
```

```python
# The status of the solution is printed to the screen
print("Status:", LpStatus[prob.status])
# Output= 
# Status: Optimal

# Each of the variables is printed with it's resolved optimum value
for v in prob.variables():
    print(v.name, "=", v.varValue)
# Output=
# Medicine_1_units = 3.0
# Medicine_2_units = 4.0

# The optimised objective function value is printed to the screen
print("Total Health that can be restored = ", value(prob.objective))
# Output= 
# Total Health that can be restored =  155.0
```
